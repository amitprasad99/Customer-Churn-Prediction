```python
ip=churn.drop('Exited',axis=1)
op=churn.Exited
```

```python
ip
```

```python
op
```

```python
from sklearn.model_selection import train_test_split
xtr,xt,ytr,yt=train_test_split(ip,op,train_size=0.8)
```

```python
xtr
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
xtr=sc.fit_transform(xtr)
xt=sc.fit_transform(xt)
```

```python
xtr
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(xtr,ytr)
```

```python
y_pred=lr.predict(xt)
y_pred
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score,f1_score
acc=accuracy_score(y_pred,yt)
rec=recall_score(y_pred,yt,average='macro')
pre=precision_score(y_pred,yt,average='macro')
f1=f1_score(y_pred,yt,average='macro')

print(acc,rec,pre,f1)
```

```python
#GNB
from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB()
gnb.fit(xtr,ytr)
```

```python
g_pred=gnb.predict(xt)
g_pred
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score,f1_score
acc=accuracy_score(g_pred,yt)
rec=recall_score(g_pred,yt,average='macro')
pre=precision_score(g_pred,yt,average='macro')
f1=f1_score(g_pred,yt,average='macro')

print(acc,rec,pre,f1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(xtr,ytr)
```

```python
k_pred=knn.predict(xt)
k_pred
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score,f1_score
acc=accuracy_score(k_pred,yt)
rec=recall_score(k_pred,yt,average='macro')
pre=precision_score(k_pred,yt,average='macro')
f1=f1_score(k_pred,yt,average='macro')

print(acc,rec,pre,f1)
```

```python
from sklearn.svm import SVC  #SVC-->Support Vector Classifier
svc=SVC(kernel='rbf',C=10000,gamma=0.01)  #C=Regularization Parameter
svc.fit(xtr,ytr)
```

```python
s_pred=svc.predict(xt)
s_pred
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score,f1_score
acc=accuracy_score(s_pred,yt)
rec=recall_score(s_pred,yt,average='macro')
pre=precision_score(s_pred,yt,average='macro')
f1=f1_score(s_pred,yt,average='macro')

print(acc,rec,pre,f1)
```
